```python
#gTTS: 구글에서 지원하는 파이썬의 음성 변환 모듈
#pip install gtts
from gtts import gTTS
from IPython.display import Audio
from IPython.core.display import display

#gTTS로 텍스트를 한국어(lang='ko')음성으로 변환 한다.
def aivoice(text):
    tts = gTTS(text=text, lang='ko')
    file = 'wav/voice.mp3'
    tts.save(file)
    display(Audio(file, autoplay=True))

aivoice(input('text를 입력하세요:'))
```

```python
with open('aiTest.txt','r', encoding='UTF8') as f:
    text_file = f.read()
    print(text_file)

aivoice(text_file)
```
