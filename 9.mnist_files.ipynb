```python
#Dataset : https://www.kaggle.com/datasets/sripaadsrinivasan/audio-mnist
import numpy as np
import pandas as pd
import os

ROOT_DIR = 'c:/data/AudioMnist'
annot_list = []

for folder in os.listdir(ROOT_DIR):
    if os.path.isdir(ROOT_DIR + '/' + folder):  #파일이 아닌 폴더만 찾는다.
        for file in os.listdir(ROOT_DIR + '/' + folder):
            file_path = ROOT_DIR + '/' + folder + '/' + file
            label = file[0]
            annot_list.append((file_path, label))
```

```python
annot_list[200]
```

```python
import random
random.shuffle(annot_list)   #전체파일 리스트 무작위 섞기

annot_list[200]
```

```python
#학습용, 검증용 Data 나누기
train_size = int(len(annot_list)*0.75)
train_list = annot_list[:train_size]
test_list = annot_list[train_size:]
```

```python
test_list
```

```python
#최종 csv파일 생성
import csv

header_list = ['file_name', 'label']
with open(ROOT_DIR+'/train_audioMNIST.csv', mode='w', newline='') as csv_file:  
    csv_writer = csv.writer(csv_file)
    csv_writer.writerow(header_list) 
    for item in train_list:
        csv_writer.writerow([item[0], item[1]])  

with open(ROOT_DIR+'/test_audioMNIST.csv', mode='w', newline='') as csv_file:
    csv_writer = csv.writer(csv_file)
    csv_writer.writerow(header_list) 
    for item in test_list:
        csv_writer.writerow([item[0], item[1]])
```
